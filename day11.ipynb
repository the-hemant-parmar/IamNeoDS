# Exploratory Data Analysis (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## For 'Iris' dataset

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.pairplot(df)
plt.grid(True)
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation, cmap='Spectral')
plt.title('Correlation between flower properties')
plt.show()
```

```python
# fig, axes =  plt.subplots(nrows=1,ncols=2)
plt.figure(figsize=(10,8))
sns.regplot(x=df['petal_length'], y=df['petal_width']) # since the correlation between them is quite high
plt.title("Dependence petal length on width")
plt.xlabel("Petal lnegth")
plt.ylabel("Petal Width")
plt.grid()
plt.show()
```

```python
# Our target variable is string, we should convert it to numerical value
df.head()
```

```python
mapping = {'setosa':0, 'virginica':1, 'versicolor':2}
df['species_id'] = df['species'].map(lambda x:mapping[x])
```

```python
df
```

```python
df2 = df.drop(columns='species')
df2.head()
```

```python
correlation2 = df2.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation2, cmap="Spectral")
plt.title("Correlation between flower properties")
plt.show()
```

## For 'Tips' dataset

```python
dft = sns.load_dataset('tips')
dft.head()
```

```python
sns.pairplot(dft)
plt.grid()
plt.show()
```

```python
tips_corr = dft.corr(numeric_only=True)
```

```python
plt.figure(figsize=[8,6])
sns.heatmap(tips_corr, cmap='coolwarm', fmt=".2f")
plt.title('Correlation for tipping')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x=dft['total_bill'], y=dft['tip'])
plt.title('Dependence of tip on total bill')
plt.xlabel('Total bill')
plt.ylabel('Tip given')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x="total_bill", y="tip", data=dft)
plt.title("Dependence of tip on total bill")
plt.xlabel("Total bill")
plt.ylabel("Tip given")
plt.grid()
plt.show()
```
