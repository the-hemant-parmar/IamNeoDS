Data Visualization using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3, 4, 5]
Y = [2, 3, 5, 7, 11]
```

```python
plt.plot(X, Y, marker = 'o')
plt.title('Basic line plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.show()
```

```python
plt.plot(X, Y, marker = 's')
plt.title('Basic line plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.show()
```

```python
env = {'Nitrogen':78,'Oxygen':21,'Helium':0.5,'Others':0.5}
plt.pie(env.values(), labels=env.keys())
plt.title('Gases in atmosphere')
plt.show()
```

```python
env = {'Nitrogen':78,'Oxygen':21,'Helium':0.5,'Others':0.5}
plt.pie(env.values(), labels=env.keys(), autopct="%1.2f%%")
plt.title('Gases in atmosphere')
plt.show()
```

```python
plt.scatter(X, Y, marker='o' )
plt.title('Basic Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.show()
```

```python
plt.scatter(X, Y, marker='x',c =['black','green','red','blue', 'yellow'])
# plt.scatter(X, Y, marker='x',cmap =['red','blue', 'yellow'])
plt.title('Basic Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.show()
```

```python
plt.bar(X,Y)
plt.title('Basic Bar Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(axis='y')
plt.show()
```

```python
def my_bar(x, y, label="Basic Bar Plot"):
    plt.bar(x, y)
    plt.title(label)
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.grid(axis="y")
    plt.show()
```

```python
my_bar(X, Y)
```

```python
my_bar(env.keys(), env.values(), 'Gases in the atmosphere')
```
